```python
# load jsonl data from a file
import torch
import json
from pprint import pprint
file_path = '../../data/prover-llm_v0/workbook_policy.jsonl'
policy_data = []
with open(file_path, 'r') as f:
    for line in f:
        policy_data.append(json.loads(line))

file_path = '../../data/prover-llm_v0/workbook_value.jsonl'
value_data = []
with open(file_path, 'r') as f:
    for line in f:
        value_data.append(json.loads(line))

file_path = '../../data/prover-llm_v0/comments/comments_v2.txt'
comments = []
with open(file_path, 'r') as f:
    for line in f:
        comments.append(line.strip())

for i in range(10):
    print(policy_data[i]['prompt'])
    print(value_data[i]['value'])
    policy = policy_data[i]['logits']
    policy = torch.tensor(policy)
    policy = torch.softmax(policy, dim=-1)

    # sort the comments by policy, and then print in order of policy
    comment_and_policy = sorted(zip(comments, policy), key=lambda x: x[1])
    for comment, policy in comment_and_policy[::-1][:20]:
        print(round(policy.item(), 4), comment)
```
